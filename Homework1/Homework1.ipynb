## I am going to predict if an employee is city resident or not based on age and years in service 

## The data seems to be incomplete for both *AGE* and *YEARS_IN_SERVICE* attributes. Lets fix that first. Replacing the missing values with *median* for both the attributes

```python
df['AGE'].head(10)
```

```python
age_median = df['AGE'].median()
```

```python
df['AGE'] = df['AGE'].fillna('U')
```

```python
df['AGE'] = df['AGE'].replace('U',age_median)
```

```python
df['AGE'].head(10)
```

```python
df['YEARS_IN_SERVICE'].head(10)
```

```python
yis_median = df['YEARS_IN_SERVICE'].median()
```

```python
df['YEARS_IN_SERVICE'] = df['YEARS_IN_SERVICE'].fillna('U')
```

```python
df['YEARS_IN_SERVICE'] = df['YEARS_IN_SERVICE'].replace('U',yis_median)
```

```python
df['YEARS_IN_SERVICE'].head(10)
```

## Now we can try using these to predict if an employee is a city resident of not

## But to predict that, we need to have CITY_RESIDENT? attribute as numeric type, which we will convert to a numeric type and put it in a new attribute

```python
classmap = {}
for idx, name in enumerate(df['CITY_RESIDENT?'].unique()):
    classmap[name] = float(idx)
    print(idx, name)
df['CITY_RESIDENT'] = df['CITY_RESIDENT?'].apply(lambda x: classmap[x])
```

## The below scatter plot shows the distribution of AGE vs YEARS IN SERVICE. I believe these two attributes may be related and help in understanding if the employees live inside the city or not

```python
df.plot.scatter(x = 'AGE', y = 'YEARS_IN_SERVICE')
```

## Lets add the CITY_RESIDENT attribute to the above plot. As we are trying to predict whether an employee is CITY RESIDENT or not, we only have two values to predict (Yes/No). In the below plot 1 or yellow in colormap mean the person is a city resident and 0 or Blue means otherwise

## We can infer from the below distribution and slight distinction that employees in their early years of service tend to live in city rather than outside. But at the same time, majority of the employees still live outside the city. Also, it is observable that elder employees who have spent many years in service tend to live outside the city more. This all inference is assuming the dataset only contains the employees who are employed in the city.

```python
df.plot.scatter(x = 'AGE', y = 'YEARS_IN_SERVICE', c = 'CITY_RESIDENT', colormap='plasma')
```

## Below I have also tried using the pair of ASSIGNMENT and YEARS_OF_SERVICE as the predictor variables for CITY_RESIDENT? prediction, because ASSIGNMENT may cause employees to live near or far from the city along with YEARS_IN_SERVICE as the years in service pass by, employee understand the working environment.
